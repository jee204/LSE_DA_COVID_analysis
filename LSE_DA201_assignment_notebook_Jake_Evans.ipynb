```python
cov_vac = cov_vac[['Province/State','Date','Deaths','Cases','Recovered',
                    'Hospitalised', 'First Dose', 'Second Dose']]
```

```python
# Groupby and calculate difference between first and second dose
```

```python
# A subset is created of the grouped Provinces/States, which are aggregated by sum
cov_vac_provstate = cov_vac.groupby(['Province/State']).agg('sum').reset_index()

# A column is created for indviduals who only received their first vaccine dose
cov_vac_provstate['First Dose Only'] = cov_vac_provstate['First Dose'] - cov_vac_provstate['Second Dose']

# The table has been sorted by first dose only individuals in ascending order.
cov_vac_provstate.sort_values(by=['First Dose Only'], inplace=True)

print(cov_vac_provstate)
```

```python
# Groupby and calculate the difference between first and second dose over time
```

```python
# As first dose and second dose columns is daily quantity rather than quantity. A column was added for both that cumalates
# over time.

# Group by Date
cov_vac_date = cov_vac.groupby(['Date']).agg('sum').reset_index()


# Create cumalative columns of vaccines filling initially with 0 to begin loop
cov_vac_date['First Dose Cumalative'] = 0
cov_vac_date['Second Dose Cumalative'] = 0

for i in cov_vac_date.index:
    
# The first value in column is assigned value of 0 to begin running total
    if i == 0:
        cov_vac_date['First Dose Cumalative'][i] = 0
        cov_vac_date['Second Dose Cumalative'][i] = 0
        
# Each value sums the running total and daily vaccine quantity        
    else:
        cov_vac_date['First Dose Cumalative'][i] = cov_vac_date['First Dose Cumalative'][i-1] + cov_vac_date['First Dose'][i] 
        cov_vac_date['Second Dose Cumalative'][i] = cov_vac_date['Second Dose Cumalative'][i-1] + cov_vac_date['Second Dose'][i] 


# A first dose only cumalative column was created.
cov_vac_date['First Dose Only Cumalative'] = cov_vac_date['First Dose Cumalative'] - cov_vac_date['Second Dose Cumalative']

# Print new DataFrame to sense check data
print(cov_vac_date)
```

```python
# A subset was created to allow lineplots to be created for deaths and recovered numbers over time for the different regions.
cov_vac_deaths_and_recovered = cov_vac[['Province/State', 'Date', 'Cases', 'Deaths', 'Recovered']]
cov_vac_deaths_and_recovered['Date'] = cov_vac_deaths_and_recovered['Date'].apply(pd.Timestamp)

# Creates a list of the provinces to use for a loop
provinces = list(cov_vac_provstate['Province/State'])

# This loop plots Deaths over time for each region in DataFrame. From this we can see most regions experience peaks, however
# the Isle of Mann, Cayman Islands, Channel Islands and Turks & Caicos Islands appear to continue increasing in deaths.

for p in provinces:

# Creates a subset which filters the data by a certain province in the loop
    graph = cov_vac_deaths_and_recovered[cov_vac_deaths_and_recovered['Province/State'] == p]

# This subset is then group by Date and aggregated by sum
    graph = graph.groupby(['Date']).agg('sum').reset_index()

# This graph is then plotted.
    sns.lineplot(x = 'Date', y = 'Deaths', data=graph, ci=None, label='Deaths')
    plt.axvline(pd.Timestamp('2021-01-11'),color='r', linestyle='dashed', label= 'First Vaccine Introduction')
    plt.xticks(rotation=90)
    plt.title(f'{p} Deaths over time')
    plt.legend()
    plt.savefig(f'Deaths_over_Months_Line_Graph_{p}.png')
    plt.show()
```

```python
# Import the tweet data set

tweets = pd.read_csv('tweets.csv')
```

```python
# Explore the data: info(), head()
print(tweets.head())
print(tweets.info())
```

```python
# Explore the structure, count the tweets, get the elements of interest
# There are 3,960 rows and 21 columns

print(tweets.shape)
```

```python
# The aim is to extract and count hashtags from tweets. From the head function, it can be seen tweets fall in the text column. 

# The text column is changed from object datatype to string to make extracting hashtags easier
tweets['text'] = tweets['text'].astype(str)

# A DataFrame is created of the text column with only tweets containing a hashtag.
tweets_hashtag = tweets[tweets['text'].apply(lambda x: "#" in x.lower())]

print(tweets_hashtag['text'])
```

```python
# This line of code prints the ammount of tweets that contain a hastag, 2,889 of 3,960 tweets contain a hashtag.
tweet_count = tweets_hashtag['text'].count()
print(f'Number of tweets with #s: {tweet_count}')

```

```python
# The text column containing hashtags is transformed to a list and assigned the alias tweet_list

tweets_hashtag = tweets_hashtag['text']

tweet_list = tweets_hashtag.values.tolist()
```

```python
# This loop extracts the hashtags from the previously created list and is added to a new list.

# k is set as an index list to loop through the length of the hashtag text list
l = len(tweet_list) - 1
k = list(range(0, l))

# All hashes is set as an empty list to store hashtags
all_hashes = []

# The loop goes through each value in the text list array
for i in k:
    
# This loop then looks at each word in the text string   
    for n in tweet_list[i].split():

# If a word in the text string starts with a # it is added to the array all_hashes        
        if n.startswith("#"):
            all_hashes.append(n)
            
# This displays how many hashtags were collected in total, there are 13,660
print(len(all_hashes))

# The first 30 hashtags from the list are displayed
print(all_hashes[:30])
```

```python
# The following loop takes the list of hashtags and removes the initial '#'

ht = '#'
hash_word = []

for t in all_hashes:
    s = t.lstrip(ht)
    hash_word.append(s)
    
# This loop changes each ammeded hashtag into lowercase, will allow easier groupings of hashtags which have the same sentiment
# yet differences in capitalised letters i.e. 'Covid' and 'COVID'
for i in range(len(hash_word)):
    hash_word[i] = hash_word[i].lower()
```

```python
# The ammended hashtags are then transformed into a DataFrame called 'Hashtags' with a column named 'Tag'
Hashtags = pd.DataFrame(hash_word)
Hashtags.columns = ['Tag']

# A second column is added to the DataFrame called count with every value being 1, this is used to count the hashtags 
# once grouped.
Hashtags['Count'] = 1

# Hashtags is then grouped by 'Tag', with the count column aggreagted by sum. As every row has a 'Count' value of 1, this 
# should sum to the total times each tag was used once grouped.
Hashtags = Hashtags.groupby(['Tag']).agg('sum').reset_index()

# The DataFrame is then sorted in descending order. Once printed it is clear that 'covid19' is the most common hashtag used.
Hashtags.sort_values(by=['Count'], inplace=True, ascending=False)
print(Hashtags)
```

```python
# Subsets are created containing different key words such as 'covid, 'corona' and 'vac' (used for vaccine or vaccination).
covid = Hashtags[Hashtags['Tag'].str.contains('covid')]
corona = Hashtags[Hashtags['Tag'].str.contains('corona')]
vaccination = Hashtags[Hashtags['Tag'].str.contains('vac')]

# A subset codvid_hashtags is created combining those previous subsets to give a dataframe of just the keywords
covid_hashtags = pd.concat([covid, corona, vaccination], axis=0)

# This line removes duplicates, hashtags that contain several key words such as'covidvaccination
covid_hashtags = covid_hashtags.drop_duplicates () 

# This new DataFrame is then sorted into descending order
covid_hashtags.sort_values(by=['Count'], inplace=True, ascending=False)

# By filtering the data to hashtags with a count above 80, the top 10 hashtags remain which is a suitable amount for a bar graph
Hashtag_Count_Graph = covid_hashtags[covid_hashtags['Count'] > 80] 
 
print(Hashtag_Count_Graph)
```

```python
# Plotting aand saving a barchart of the most used Covid hashtags
sns.barplot(x='Tag', y='Count', data=Hashtag_Count_Graph, palette='colorblind').set(title='Count of COVID related hashtags')
plt.savefig('Count of COVID related hashtags.png')
```

```python
# A subset of the data for the Channel Islands is created to use for the Moving Average function
channel_islands = cov_ci[cov_ci['Province/State'] == "Channel Islands"] 
channel_islands = channel_islands[['Hospitalised']]
channel_islands.groupby(['Date']).agg('sum').reset_index()
```

```python
# Define function to plot moving averages
def plot_moving_average(series, window, plot_intervals=False, scale=1.96):
    
    # Create a rolling window to calculate the rolling mean using the series.rolling function
    rolling_mean = series.rolling(window=window).mean()
    
    # Declare the dimensions for the plot, plot name and plot the data consisting of the rolling mean from above 
    plt.figure(figsize=(18,4))
    plt.title('Moving average\n window size = {}'.format(window))
    plt.plot(rolling_mean, 'g', label='Simple moving average trend')

    
    # Plot confidence intervals for smoothed values
    if plot_intervals:
        
        # Calculate the mean absolute square 
        mae = mean_absolute_error(series[window:], rolling_mean[window:])
        
        # Calculate the standard deviation using numpy's std function
        deviation = np.std(series[window:] - rolling_mean[window:])
        
        # Calculate the upper and lower bounds 
        lower_bound = rolling_mean - (mae + scale * deviation)
        upper_bound = rolling_mean + (mae + scale * deviation)
        
        # Name and style upper and lower bound lines and labels 
        plt.plot(upper_bound, 'r--', label='Upper bound / Lower bound')
        plt.plot(lower_bound, 'r--')
    
    # Plot the actual values for the entire timeframe
    plt.plot(series[window:], label='Actual values')
    plt.legend(loc='best')
    plt.grid(True)
```

```python
# Define function to calculate the mean absolute error
def mean_absolute_error(a, b): return abs(b - a)
```

```python
# Use the moving average function for the Channel Islands with a 7 day window
plot_moving_average(channel_islands, 7, plot_intervals=True)

# The moving average function works and helps visualise the trend of hospitalised individuals whilst keeping it smooth by 
# reducing volatility with a moving average. We can see using a 7 day window, there is not a huge difference between actual 
# and mean values, day to day there isn't much volatility. The consultant has designed this function to try and help predict 
# increases in hospitalised individuals to give more preparation time.
```
